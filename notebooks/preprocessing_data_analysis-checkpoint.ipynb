```python
import pandas as pd
import matplotlib.pyplot as plt
import os

file_path = "../output/"
# List of CSV files - you can modify this list to add your specific files
csv_files = ['inst_score_train.csv', 'pos_score_train.csv', 'cc_score_train.csv', 'bubl_score_train.csv', 'train_house_score.csv']

# Prepare the subplot environment
fig, axes = plt.subplots(nrows=len(csv_files), ncols=1, figsize=(10, 5 * len(csv_files)))

# Check if there is only one plot, if so, wrap axes in a list for easy iteration
if len(csv_files) == 1:
    axes = [axes]

# Iterate over the list of files and their corresponding axes
for ax, file in zip(axes, csv_files):
    # Read each file
    df = pd.read_csv(os.path.join(file_path, file))
    
    # Extract the score column, assuming the naming convention ends with '_score'
    # This extracts the last column that ends with '_score'
    score_column = [col for col in df.columns if 'score' in col][-1]
    
    # Plot a histogram of the score column
    df[score_column].plot(kind='hist', ax=ax, bins=30, alpha=0.7, color='blue')
    
    # Set the title to the file name for identification
    ax.set_title(f'Distribution of {score_column} in {file}')
    ax.set_xlabel(score_column)
    ax.set_ylabel('Frequency')

# Adjust layout
plt.tight_layout()
plt.show()
```
